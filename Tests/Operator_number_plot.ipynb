```python
data0 = pd.read_csv("../level_curve_dt/1.2[0,0.1]_6p_10ecl_pois_var_1cus.csv")
data1 = pd.read_csv("../level_curve_dt/1.2[0,0.1]_6p_30ecl_pois_var_1cus.csv")
data2 = pd.read_csv("../level_curve_dt/1.2[0,0.1]_6p_50ecl_pois_var_1cus.csv")
probs = [0,0.2,0.4,0.6,0.8,1]
data_lists0 = [0]*len(probs); data_lists1 = [0]*len(probs); data_lists2 = [0]*len(probs)
lams = [data0.fix_lam[0],data1.fix_lam[0],data2.fix_lam[0]]
legs = [f'scale = {lam}' for lam in lams]
colors = ['b','g','r','m','c']
if len(probs) > len(colors):
    styles = [f'{c}o' for c in colors]
    styles += ['o']*(len(probs)-len(colors))
else:
    styles = [f'{c}o' for c in colors[:len(probs)]]
    
for i in range(len(probs)):
    data_lists0[i] = data0[data0.pause_prob==probs[i]]
    data_lists1[i] = data1[data1.pause_prob==probs[i]]
    data_lists2[i] = data2[data2.pause_prob==probs[i]]
    
    plot_info=[ [[data_lists0[i].eps, data_lists0[i].avg_wt],styles[0],legs[0]],
          [[data_lists1[i].eps, data_lists1[i].avg_wt],styles[1],legs[1]],
          [[data_lists2[i].eps, data_lists2[i].avg_wt],styles[2],legs[2]]  ]

    title = f'Pause probability = {probs[i]}'
    input_label = None
    output_label = 'avg. waiting time (hour)'

    plot_2D(plot_info,title,input_label,output_label,bloc='best')
```

## Operator vs. Avg. waiting time

```python
data0 = pd.read_csv("../level_curve_dt/1.2[0,0.3]_2p_10ecl_pois_var_1cus.csv")
data1 = pd.read_csv("../level_curve_dt/1.2[0,0.3]_2p_30ecl_pois_var_1cus.csv")
data2 = pd.read_csv("../level_curve_dt/1.2[0,0.1]_6p_50ecl_pois_var_1cus.csv")
probs = [0,1]
scales = [10, 30, 50]
data = [data0, data1, data2]
lams = [data0.fix_lam[0],data1.fix_lam[0],data2.fix_lam[0]]
legs = [f'pause_prob = {prob}' for prob in probs]
colors = ['b','g','r','m','c']
styles = ['bo','ro']
    
for i in range(len(scales)):
    data00 = data[i][data[i].pause_prob==0]
    data00 = data00[data00.avg_wt < 40]
    data01 = data[i][data[i].pause_prob==1]
    data01 = data01[data01.avg_wt < 40]

    
    plot_info=[ [[data00.avg_wt, data00.op_num],styles[0],legs[0]],
          [[data01.avg_wt, data01.op_num],styles[1],legs[1]],
          [[[5,5], [np.mean([min(data00.op_num),min(data01.op_num)])-0.5,
                    np.mean([max(data00.op_num),max(data01.op_num)])+0.5]],'g:','objective: 5 hr']
           ]

    title = f'Job scale = {scales[i]}'
    input_label = 'avg. waiting time (hour)'
    output_label = 'operator number'

    plot_2D(plot_info,title,input_label,output_label,bloc='best')
```
