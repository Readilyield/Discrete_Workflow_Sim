## Running $\kappa$ = 0.5, 1, 2, 4
## fixing op_num, change job avg. length and scale

```python
k = 5
path = "../ecldata/ecldata_without_float.csv"
c, f_lam, f_rls = get_multiplier(k, tol = 1e-3, check = True, path = path)
jobs, jobs_prob, mean_ot, total_ot, total_jobs = csv_to_jobs(path, mul=c)
'''rounding up/down lambda acceptable?'''
print(mean_ot+1)
#op_num = (1+eps)*f_lam*(cus_num*1+cus_num*mean_ot)/24

op_nums = csv_to_data("level_curve_dt/1.0ot_op_10ecl_pois_var_1cus.csv",attr='op_num')

tmp = []
epss = np.linspace(0.3,0,31) 
#0.5-0, 26 discrete scales (0.01)
for eps in epss:
    op_num_tmp = (1+eps)*f_lam*1*(mean_ot+1)/24
    tmp.append(round(op_num_tmp,2))
    
error = max(abs(np.array(tmp)-np.array(op_nums)))
print(f'op_num error: {error}')
test_batch_wpic(jobs, jobs_prob, mean_ot, total_ot, total_jobs, 
                epss, time_span = 300, f_lam = f_lam, f_rls = f_rls, 
                poisson = True, var_op = True, 
                cus_num=1, pause_probs = [0,0.5,1], code=f'12.10_{k}ot_EFop', 
                runs = 20, plots = [True, False, False, False])
```
