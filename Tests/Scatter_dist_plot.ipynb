## Empirical level curves-Plot
Suppose we have 
$$f(s, \epsilon) = t$$
where s = scale (num. of customer or num of incoming jobs/day),
<br>
$\epsilon$ = HO surplus, s.t. HO = (1+$\epsilon$)HI, 
<br>
and t = mean waiting time (hours)
\begin{align}
\text{Want to find } C_f(t) = \{(s, \epsilon) | f(s, \epsilon) = t\}
\\
\\
\text{Moreover, want to find a function } g_t \;\text{ s.t. } C_f(t) = \{(s, g_t(s))\}
\end{align}

```python
#data = pd.read_csv("test1_determ_df.csv") #if .csv is in the same folder as this notebook
data = pd.read_csv("level_curve_dt/test1_determ_df.csv")
data_list_0 = [];data_list_1 = []

#mean_wait time = 0-1 hr
data_list_0.append(data[(data.avg_wt<=1)&(data.pause_prob==0)])
data_list_1.append(data[(data.avg_wt<=1)&(data.pause_prob==1)])
#mean_wait time = 1-5 hr
data_list_0.append(data[(data.avg_wt>1)&(data.avg_wt<=5)&(data.pause_prob==0)])
data_list_1.append(data[(data.avg_wt>1)&(data.avg_wt<=5)&(data.pause_prob==1)])
#mean_wait time = 5-20 hr
data_list_0.append(data[(data.avg_wt>5)&(data.avg_wt<=20)&(data.pause_prob==0)])
data_list_1.append(data[(data.avg_wt>5)&(data.avg_wt<=20)&(data.pause_prob==1)])
#mean_wait time 20-50 hr
data_list_0.append(data[(data.avg_wt>20)&(data.avg_wt<=50)&(data.pause_prob==0)])
data_list_1.append(data[(data.avg_wt>20)&(data.avg_wt<=50)&(data.pause_prob==1)])
#mean_wait time 50-100 hr
data_list_0.append(data[(data.avg_wt>50)&(data.avg_wt<=100)&(data.pause_prob==0)])
data_list_1.append(data[(data.avg_wt>50)&(data.avg_wt<=100)&(data.pause_prob==1)])
#mean_wait time >100 hr
data_list_0.append(data[(data.avg_wt>100)&(data.pause_prob==0)])
data_list_1.append(data[(data.avg_wt>100)&(data.pause_prob==1)])

#data containers for plotting
plot_info_0 = [];plot_info_1 = []
styles = ['bo','go','yo','ro','mo','ko']
legs = ['[0,1]','(1,5]','(5,20]','(20,50]','(50,100]','(100,']
#plot_info.append([[data.eps,data.cus_num],'bo',''])
for i in range(len(data_list_0)):
    plot_info_0.append([[data_list_0[i].eps, data_list_0[i].cus_num],styles[i],legs[i]])
    plot_info_1.append([[data_list_1[i].eps, data_list_1[i].cus_num],styles[i],legs[i]])
```

```python
title_0 = 'Avg. Wait Time Full Dist., P=0'
title_1 = 'Avg. Wait Time Full Dist., P=1'
input_label = 'op. surplus %'
output_label = 'num. avg. daily job'

plot_2D(plot_info_0,title_0,input_label,output_label,bloc='upper left')
plot_2D(plot_info_1,title_1,input_label,output_label,bloc='upper left')
```
